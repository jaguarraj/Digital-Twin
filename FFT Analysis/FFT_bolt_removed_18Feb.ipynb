```python
import lvm_read
import numpy as np
import pandas as pd
from scipy.stats import *
from scipy import pi
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import time
from collections import Counter
from numpy.fft import rfft, rfftfreq
```

```python
one_Bolt_file = 'C:/Users/intel/Desktop/Mtech_Proj/Digital twin_29 Jan 2021/dataset/Bolt1_Removed.lvm'
three_Bolt_file = 'C:/Users/intel/Desktop/Mtech_Proj/Digital twin_29 Jan 2021/dataset/Bolt1_3_Removed.lvm'
Healthy_file = 'C:/Users/intel/Desktop/Mtech_Proj/Digital twin_29 Jan 2021/dataset/Healthy.lvm'

lvm_one_Bolt = lvm_read.read(one_Bolt_file, read_from_pickle=False)
lvm_three_Bolt = lvm_read.read(one_Bolt_file, read_from_pickle=False)
lvm_Healthy = lvm_read.read(Healthy_file, read_from_pickle=False)
```

```python
one_Bolt_data_vib = lvm_one_Bolt[0]['data'][:,3]## Vibration data
three_Bolt_data_vib = lvm_three_Bolt[0]['data'][:,3]## Vibration data
Healthy_data_vib = lvm_Healthy[0]['data'][:,3]
```

```python
one_Bolt_data_vib = pd.DataFrame(one_Bolt_data_vib)
three_Bolt_data_vib = pd.DataFrame(three_Bolt_data_vib)
Healthy_data_vib = pd.DataFrame(Healthy_data_vib)
```

```python
#FFT
N = one_Bolt_data_vib.shape[0] #data size=3600 data pts
f_s = 2000 #Sampling frequency
dt = 1.0 / f_s #inverse of sampling rate

y = one_Bolt_data_vib.values.flatten() #to convert DataFrame to 1D array
    #acc value must be in numpy array format for half way mirror calculation
fft=rfft(y)*dt
freq=rfftfreq(N,d=dt)
FFT=abs(fft)
plt.figure(figsize=(16,10))
plt.plot(freq,FFT)
plt.title('FFT One Mounting Bolt Missing')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Vibration (g)')
```

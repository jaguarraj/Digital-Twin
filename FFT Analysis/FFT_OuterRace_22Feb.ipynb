```python
import lvm_read
import numpy as np
import pandas as pd
from scipy.stats import *
from scipy import pi
from scipy.fftpack import fft
import matplotlib.pyplot as plt
import time
from collections import Counter
from numpy.fft import rfft, rfftfreq
```

```python
OR_file = 'C:/Users/intel/Desktop/Mtech_Proj/Digital twin_29 Jan 2021/dataset/Vib_Pow_OutterRace.lvm'
Healthy_file = 'C:/Users/intel/Desktop/Mtech_Proj/Digital twin_29 Jan 2021/dataset/Healthy.lvm'

lvm_OR = lvm_read.read(OR_file, read_from_pickle=False)
lvm_Healthy = lvm_read.read(Healthy_file, read_from_pickle=False)
```

```python
OR_data_vib = lvm_OR[0]['data'][:,1]## Vibration data
Healthy_data_vib = lvm_Healthy[0]['data'][:,3]
```

```python
OR_data_vib = pd.DataFrame(OR_data_vib)
Healthy_data_vib = pd.DataFrame(Healthy_data_vib)
```

```python
#FFT
N = OR_data_vib.shape[0] #data size=3600 data pts
f_s = 2000 #Sampling frequency
dt = 1.0 / f_s #inverse of sampling rate

y = OR_data_vib.values.flatten() #to convert DataFrame to 1D array
    #acc value must be in numpy array format for half way mirror calculation
fft=rfft(y)*dt
freq=rfftfreq(N,d=dt)
FFT=abs(fft)
plt.figure(figsize=(10,8))
plt.plot(freq,FFT)

plt.title('FFT OuterRace Fault Bearing')
plt.xlabel('Frequency (Hz)')
plt.ylabel('Vibration (g)')
```
